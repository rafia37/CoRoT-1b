```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits, ascii
import pdb, mpld3, sys
from mpld3 import plugins
sys.path.insert(0, '../')
from WFC3Dispersion import WFC3Dispersion

plt.rcParams['figure.figsize'] = (12, 6)
mpld3.enable_notebook()
```

```python
fnames  = ascii.read('img_filenames.csv')

img_list = []
hdr_list = []
for fn in fnames:
    hdu = fits.open(fn[0])
    header = hdu[0].header
    if header['OBSTYPE'] == 'IMAGING':
        image = hdu[1].data
        img_list.append(image)
        hdr_list.append(header)
    hdu.close()
```

```python
fig, ax = plt.subplots()
ax.imshow(img_list[0])
plugins.connect(fig, plugins.MousePosition(fontsize=14))
```
